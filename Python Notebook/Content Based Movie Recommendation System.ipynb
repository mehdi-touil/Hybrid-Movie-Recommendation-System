```python
pd.set_option('display.max_columns',100)
movie = pd.read_csv('../dataset/tmdb_5000_movies.csv')
credit = pd.read_csv('../dataset/tmdb_5000_credits.csv')
```

```python
import ast #The ast module helps Python applications to process trees of the Python abstract syntax grammar
#eval: This is very powerful, but is also very dangerous if you accept strings to evaluate from untrusted input. 
#Suppose the string being evaluated is "os.system('rm -rf /')" ? It will really start deleting all the files on your computer.

#ast.literal_eval: Safely evaluate an expression node or a string containing a Python literal or container display. 
#The string or node provided may only consist of the following Python literal structures: strings, bytes, numbers, tuples, 
#lists, dicts, sets, booleans, None, bytes and sets.
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
movie.dropna(inplace=True)
```

```python
movie['genres'] = movie['genres'].apply(convert)
movie.head()
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
movie.head(2)
```

```python
movie['cast']
```

```python
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movie['cast'] = movie['cast'].apply(convert)
movie.head()
```

```python
movie['cast'] = movie['cast'].apply(lambda x:x[0:3])
```

```python
movie.head(2)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movie['crew'] = movie['crew'].apply(fetch_director)
```

```python
movie.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movie['cast'] = movie['cast'].apply(collapse)
movie['crew'] = movie['crew'].apply(collapse)
movie['genres'] = movie['genres'].apply(collapse)
movie['keywords'] = movie['keywords'].apply(collapse)
```

```python
movie.head()
```

```python
movie['overview'] = movie['overview'].apply(lambda x:x.split())
```

```python
movie
```

```python
movie['tags'] = movie['overview'] + movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew']
```

```python
new = movie.drop(columns=['overview','genres','keywords','cast','crew'])
new
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
